```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# diagramas de dispersion (scatterplot)

x = np.arange(1,20)
y = x**2
y2 = x**3

plt.scatter(x,y,color="red", marker="*", label="x**2*") # crear diagrama
plt.scatter(x,y2,color="blue", marker="+", label="x**3") # crear diagrama
plt.title("Funcion polinomica", fontsize=20, color="black") # titulo de grafico
plt.xlabel("eje x") # nombre eje x 
plt.ylabel("eje y") # nombre eje y
plt.grid() # poner malla 
plt.legend() # ver leyenda de los datos


plt.text(2.5,5000,"texto de prueba", color="blue")
plt.plot(x, y2, color="pink", label="line")
```
