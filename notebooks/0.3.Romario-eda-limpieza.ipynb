### EDA limpieza de datos

```python
# importar librerias de manipulacion de datos
import numpy as np
import pandas as pd

# importar librerias del sistema

import os
from pathlib import Path
```

```python
# ruta del proyecto
root_dir = Path(".").resolve().parent
print("root dir:", root_dir)
```

```python
# ruta del proyecto
filename = "llamadas123_julio_2022.csv"
data_dir = "raw"
file_path = os.path.join(root_dir,"data",data_dir, filename)
print(file_path)

# C:\Users\romario.silva\Documents\ESEIT\Herramientas y Lenguajes para Big Data\BigData2022B\data\raw
```

```python
# leer archivo
data = pd.read_csv(file_path, encoding="latin-1", sep=";")


# mostrar primeras filas
data.head()
```

```python
# Eliminar los registros duplicados, dejando solo el primer registro

print("forma inicial", data.shape)

data = data.drop_duplicates()

print("forma final", data.shape)
```

```python
# esquema de la tabla

data.info()
```

```python

# realiza un conteo de los diferentes valores para una columna especifica, incluyendo los valores nulos.
data["UNIDAD"].value_counts(dropna=False)
```

```python
# Remplazar en la columna UNIDAD los nulos por SIN_DATO
data["UNIDAD"].fillna("SIN_DATO").value_counts(dropna=False)
```

```python
# Sobre escribir la columna UNIDAD que tenia valores nulos por una columna nueva sin nulos
data["UNIDAD"] = data["UNIDAD"].fillna("SIN_DATO")

data.info()
```

```python
# manipulacion de fechas

col = "FECHA_INICIO_DESPLAZAMIENTO_MOVIL"
data[col] # ver columna
data[col].min() # fecha minima
data[col].max() # fecha maxima
```

```python
# aplicar filtros

val_min = data[col].min()
data[data[col] == val_min] # traer todos los registros iguales a la variable val min

data[data[col] != val_min] # traer todos los registros diferentes a la variable val min
```

```python
# Formato de fechas (datetime)

nueva_columna = pd.to_datetime(data[col], errors="coerce") # convertir columna a tipo fecha y omitir errores con errors="coerce"
```

```python
# Sobre escribir la columna UNIDAD que tenia valores nulos por una columna nueva sin nulos
data["FECHA_INICIO_DESPLAZAMIENTO_MOVIL"] = nueva_columna

data.info() # ya se ve la columna tipo fecha
```

```python
# funcion que reciba un string, con fechas y regrese un valor en datetime usando parse

def string_to_datetime(valor_str):
    result = parse(valor_str, dayfirst=False)
    return result


fecha_prueba = "10/05/2022"

valor_datetime = string_to_datetime(fecha_prueba)
#type(valor_datetime)
valor_datetime
```

```python

#data = data.reset_index() # resetear indices de la tabla

list_fechas = list()

n_filas = data.shape[0] # es igual a len(data["RECEPCION"])
for i in range(0,n_filas):
    
    str_fecha = data["RECEPCION"][i]

    try:
        val_datetime = string_to_datetime(valor_str=str_fecha)
        list_fechas.append(val_datetime)
    except Exception as e:
        #print(i, e)
        list_fechas.append(str_fecha)
        continue
   
    print(val_datetime)
    list_fechas
```

```python
data["RECEPCION_CORREGIDA"] = list_fechas


data["RECEPCION_CORREGIDA"] = pd.to_datetime(data["RECEPCION_CORREGIDA"], errors="coerce")

data.info()
```
